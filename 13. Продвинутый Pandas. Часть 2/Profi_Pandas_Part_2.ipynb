Задание 1

Используйте файл с оценками фильмов ml-latest-small/ratings.csv. Посчитайте среднее время жизни пользователей, которые выставили более 100 оценок. Под временем жизни понимается разница между максимальным и минимальным значением столбца timestamp для данного значения userId.

```python
import pandas as pd
```

```python
ratings = pd.read_csv('ml-latest-small/ratings.csv')
ratings.head(2)
```

```python
ratings['date'] = pd.to_datetime(ratings['timestamp'], unit = 's')
```

```python
new_ratings = ratings.groupby('userId').agg({'movieId':'count', 'date':['min', 'max']}).reset_index()#.query('movieId.count > 100')
new_ratings.head(2)
```

```python
new_ratings['ltv'] = new_ratings['date']['max'] - new_ratings['date']['min']
```

```python
ratings_for_ltv = new_ratings[new_ratings['movieId']['count'] > 100]
ratings_for_ltv['ltv'].mean()
```
